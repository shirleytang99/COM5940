```python
import requests
import csv
from bs4 import BeautifulSoup

quote_page = requests.get('http://www.alicecatherine.com/blog/') 
soup = BeautifulSoup(quote_page.content,'html.parser')
data = []
for article in soup.find_all('div',class_='cell'):
     url = article.div.h3.a.get('href') 
     post_date = article.div.time.text
     data.append((url, post_date))

with open('assignment2.csv', 'w') as csv_file: 
     writer = csv.writer(csv_file)
     header = ['url','post date'] 
     writer.writerow(header)
     for url, post_date in data: writer.writerow([url, post_date])
```
